```python
Topics :
    1. Fillna to fill missing values using different ways
    2. interpolate to make a guess on missing values
    3. dropna to drop the null values
```

```python
import pandas as pd
df = pd.read_csv("weather_data.csv", parse_dates=['day'])
# type(df.day[0])
df.head()
```

```python
df.set_index('day',inplace=True)
```

```python
df.head()
```

```python
df.isna().sum() # Checking for null values
```

```python
new_df = df.fillna(0) # Filling the null values with 8 
new_df
```

```python
# Filling the specific values to the particular column

new_df = df.fillna({
    'temperature' : 0,
    'windspeed' :0,
    'event' : 'No event'
})

new_df
```

```python
new_df = df.fillna(method='ffill', limit=1)  # Forward fill and backward fill and can limit to one place by using limit = 1
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method='time')  # Considering the time it relates it b/w the date to date and provides the data
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')  # how="all" drops the row which has all na's
new_df
```

```python
new_df = df.dropna(thresh=1)  # Keeping the row that has atleat one non NA value
new_df
```
